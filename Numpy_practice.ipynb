## Import numpy 

```
import numpy as np
```

## 1. Numpy arange() function

```
A=np.arange(1,20,2)# start=1,stop=20, step=2
print(A)
```

```
AA=np.arange(1,20,2, dtype=float)# start=1,stop=20, step=2
print(AA)
```

## 2. Creating Numpy array()

```
mylist=[[1,2,3],[2,3,4],[4,5,6]]
num_array=np.array(mylist)
print(num_array)
```

```
arr= np.array([[1,2,4],[3,1,5],[9,8,5]])
print(arr)
```

## Creating array using empty()

```
nmp=np.empty([3,2])
print(nmp)
#return uninitialized array
```

## Creating array using eye()

```
nmp1=np.eye(3,3)
print(nmp1)
## creating an 2D array with 1 in diagonals and all other elements as zero
```

## Creating array as identity()

```
nmp2=np.identity(3)
print(nmp2)
## creating an Identity element
```

## Creating array as linspace()

```
ARR=np.linspace(0,20,11)# Creating floating point array
ARR_INT=np.linspace(0,20,11, dtype= int) # Creating integer array
print(ARR);print(ARR_INT)
```

## Creating array as logspace()

```
num_arr=np.logspace(0,4,5, dtype=int)
print(num_arr)
# returns an ndarray object that contains the numbers that are evenly spaced on a log scale
```

## Creating array using zeros and ones

```
one_array=np.ones((2,3), dtype=int)
zero_array=np.zeros((2,3),dtype=int)
print(one_array);print(zero_array)
# creating numpy array of 1s and 0s
```

## The reshape function returns a new array by reshaping a new array

```
RA= np.arange(1,21,2)
Ra= RA.reshape(5,2)
print(RA);print(Ra)
```

## Creating array and fetching dtype,itemsize,ndim,shape,size

```
Raa=np.arange(1,41,2).reshape(5,4)
print(Raa);print(Raa.dtype);print(Raa.itemsize);print(Raa.ndim);print(Raa.shape);print(Raa.size)
```

## Accessing 1D ndarray object

```
B=np.arange(1,21,4)
print(B)
print(B[2])
print(B[0:4:1])## Accessing through slicing
print(B[-1:-3:-1])## Accessing through slicing
print(B>10)## Accessing using Boolean array indexing
```

## Accessing 2D ndarray object

```
C=np.arange(1,61,2).reshape(6,5)
print(C)
print(C[2][2])
print(C[1:10:2])
rowsum=C.sum(-1)
print(rowsum)
print(C[rowsum>48,:])
```
