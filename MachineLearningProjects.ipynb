```python
#Machine Learning Project: 1#
# weather forecasting #

import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
df= pd.read_csv("DailyDelhiClimateTrain.csv")
df.head()
```

```python
## Representation of mean temperature in Delhi over the Years ##
fig= px.line(df,x='date',y='meantemp',title="Mean temperature in Delhi over the years")
fig.show()
```

```python
fig= px.line(df,x='date',y='wind_speed',title="Wind Speed in Delhi over the years")
fig.show()
```

```python
fig= px.line(df,x='date',y='humidity',title="Humidity in Delhi over the years")
fig.show()
```

```python
# relationship b/n temperature and humidity #
fig= px.scatter(data_frame=df,x="humidity",y="meantemp",size="meantemp",trendline="ols",title="Humidity/Temperature")
fig.show()
```

```python
df["date"] = pd.to_datetime(df["date"], format = '%Y-%m-%d')
df['year'] = df['date'].dt.year
df["month"] = df["date"].dt.month
print(df.head())
```

```python
forecast_data = df.rename(columns = {"date": "ds", 
                                       "meantemp": "y"})
print(forecast_data)
```

```python
# Weather Forecast of Delhi #
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
model = Prophet()
model.fit(forecast_data)
forecasts = model.make_future_dataframe(periods=365)
predictions = model.predict(forecasts)
plot_plotly(model, predictions)
```
